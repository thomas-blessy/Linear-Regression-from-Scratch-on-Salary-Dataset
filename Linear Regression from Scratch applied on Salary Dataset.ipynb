```python
import matplotlib.pyplot as plt 
import pandas as pd

# Load the dataset containing salary information based on years of experience
data = pd.read_csv('Salary_Data.csv')

# Visualizing the relationship between Years of Experience and Salary
plt.scatter(data.YearsExperience, data.Salary) 
plt.xlabel('Years of Work Experience')
plt.ylabel('Salary')
plt.title('Scatter Plot of Salaries')
plt.show()

# Initialize model parameters
slope = 0          # Initial slope (m) of the best-fit line
intercept = 0      # Initial intercept (b) of the best-fit line
L = 0.0001         # Learning rate: controls how much to change the model parameters during training
epochs = 1000      # Number of iterations for gradient descent

# Gradient descent loop
for i in range(epochs):
    if i % 50 == 0:  # Print progress every 50 epochs
        print(f"Epoch: {i}")
    slope, intercept = gradient_descent(slope, intercept, data, L)

# Print the final slope and intercept after training
print(f"Final parameters: slope = {slope}, intercept = {intercept}")

# Plot the original data points and the best-fit line
plt.scatter(data.YearsExperience, data.Salary, color='black')
plt.plot(list(range(1, 11)), [slope * x + intercept for x in range(1, 11)], color="red") 
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Best Fit Line')
plt.show()
```

```python


def gradient_descent(slope_now, intercept_now, points, L):
    """Perform one step of gradient descent to update the slope and intercept."""
    
    # Initialize gradients for slope and intercept
    slope_gradient = 0
    intercept_gradient = 0
    n = len(points)  # Number of data points
    
    # Calculate gradients for each data point
    for i in range(n):
        x = points.iloc[i].YearsExperience  # Current value of Years of Experience
        y = points.iloc[i].Salary            # Current value of Salary
        
        # Calculate the predicted value (y_hat) using the current slope and intercept
        y_pred = =slope_now * x + intercept_now
        
        # Compute the gradient for the slope (m)
        # The gradient is the derivative of the loss function with respect to the slope
        slope_gradient += -(2/n) * x * (y - y_pred) #This term represents the contribution of each data point
      
        # Compute the gradient for the intercept (b)
        # The gradient is the derivative of the loss function with respect to the intercept
        intercept_gradient += -(2/n) * (y - y_pred) # This term represents the contribution of each data point
    
    # Update the slope and intercept using the calculated gradients
    slope = slope_now - slope_gradient * L  # Update rule for slope
    intercept = intercept_now - intercept_gradient * L  # Update rule for intercept
    
    return slope, intercept  # Return the updated parameters
```
